to implement the hierarchical clustering using python

STANDARD SCALAR IS USED TO MEAN AND SCALE IN THE DATASET

```python
from sklearn.preprocessing import StandardScaler
alg_1= StandardScaler()
disease = alg_1.fit_transform(disease)
disease
```

PCA IS USED TO REDUCE THE DIMENSIONALITY OF THE DATA

```python
from sklearn.decomposition import PCA
PCA=PCA(n_components=2)
alg_2=PCA.fit_transform(disease)
PCA_dis=pd.DataFrame(alg_2,columns=['p1','p2'])
PCA_dis
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(PCA_dis, method= 'ward', metric = 'euclidean'))
```

```python
y = linkage(PCA_dis, 'ward')
fig = plt.figure(figsize=(25, 10))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agg = AgglomerativeClustering(n_clusters=3)
agg.fit(PCA_dis)
```

```python
plt.scatter(PCA_dis['P1'], PCA_dis['P2'],
           c = AgglomerativeClustering(n_clusters = 3).fit_predict(PCA_dis), cmap =plt.cm.winter)
plt.show()
```
